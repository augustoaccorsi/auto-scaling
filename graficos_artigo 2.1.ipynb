```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data_xls = pd.read_excel('/content/vms.xlsx', 'Sheet1', dtype=str, index_col=None)
#data_xls = pd.read_excel('/content/netin.xlsx', 'Sheet1', dtype=str, index_col=None)
#data_xls = pd.read_excel('/content/netout.xlsx', 'Sheet1', dtype=str, index_col=None)
data_xls.to_csv('/content/csvfile.csv', encoding='utf-8', index=False) 
df=pd.read_csv('/content/csvfile.csv')

df=df.dropna()
print('Shape of data',df.shape)
#df.head()
df.tail()
```

```python
df.drop('aws1', axis =1, inplace = True)
df.drop('aws2', axis =1, inplace = True)
df.drop('aws3', axis =1, inplace = True)

df.drop('hy1', axis =1, inplace = True)
df.drop('hy2', axis =1, inplace = True)
df.drop('hy3', axis =1, inplace = True)

df.drop('cpu1', axis =1, inplace = True)
df.drop('cpu2', axis =1, inplace = True)
df.drop('cpu3', axis =1, inplace = True)

df.drop('rede1', axis =1, inplace = True)
df.drop('rede2', axis =1, inplace = True)
df.drop('rede3', axis =1, inplace = True)

df.drop('tempo', axis =1, inplace = True)
```

```python
ax = df.plot(figsize=(12,2.5))

ax.set_ylabel("Número de VMs")
ax.set_xlabel("Tempo de Execução (minutos)")

plt.show()
```

```python
data_xls = pd.read_excel('/content/cpu.xlsx', 'Sheet1', dtype=str, index_col=None)
#data_xls = pd.read_excel('/content/netin.xlsx', 'Sheet1', dtype=str, index_col=None)
#data_xls = pd.read_excel('/content/netout.xlsx', 'Sheet1', dtype=str, index_col=None)
data_xls.to_csv('/content/csvfile.csv', encoding='utf-8', index=False) 
df=pd.read_csv('/content/csvfile.csv')

df_cpu=df.dropna()
print('Shape of data',df_cpu.shape)
#df.head()
df_cpu.tail()
```

```python
data_xls = pd.read_excel('/content/net.xlsx', 'Sheet1', dtype=str, index_col=None)
#data_xls = pd.read_excel('/content/netin.xlsx', 'Sheet1', dtype=str, index_col=None)
#data_xls = pd.read_excel('/content/netout.xlsx', 'Sheet1', dtype=str, index_col=None)
data_xls.to_csv('/content/csvfile.csv', encoding='utf-8', index=False) 
df=pd.read_csv('/content/csvfile.csv')

df_net=df.dropna()
print('Shape of data',df_net.shape)
#df.head()
df_net.tail()
```

```python
#ax1 = df.plot(y=["cpu", "up_cpu", "down_cpu"])
#ax2 = df.plot(y=["net", "net_up", "net_down"])

df_cpu.drop('instancias', axis =1, inplace = True)
df_net.drop('instancias', axis =1, inplace = True)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# create dummy pandas dataframes
pd1 = df_cpu
pd2 = df_net
#create subplot figure with having two side by side plots
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2.5))
# plot first pandas frame in subplot style
pd1.plot(ax = axes[0], figsize=(12,2.5)) 
# plot second pandas frame in subplot style
pd2.plot(ax = axes[1])

plt.show()
```

```python
data_xls = pd.read_excel('/content/netcpu.xlsx', 'Sheet1', dtype=str, index_col=None)
#data_xls = pd.read_excel('/content/netin.xlsx', 'Sheet1', dtype=str, index_col=None)
#data_xls = pd.read_excel('/content/netout.xlsx', 'Sheet1', dtype=str, index_col=None)
data_xls.to_csv('/content/csvfile.csv', encoding='utf-8', index=False) 
df=pd.read_csv('/content/csvfile.csv')

df_net=df.dropna()
print('Shape of data',df_net.shape)
#df.head()
df_net.tail()
```

```python

import matplotlib.ticker as mticker


plt.figure(figsize=(12,2.5))
plt.xlabel('Tempo de Execução (minutos)')
plt.ylabel('Métricas')

ax1 = df.net.plot(grid=True, label='Rede (esquerda)')
ax1 = df.cpu.plot(grid=True, label='CPU (esquerda)')

plt.vlines(x=46, ymin=0, ymax=73, ls='--', label='Pontos onde uma VM é alocada')
plt.vlines(x=75, ymin=0, ymax=73, ls='--')
plt.vlines(x=60, ymin=0, ymax=160, ls='--', color='red', label='Pontos onde uma VM é desalocada')
plt.vlines(x=87, ymin=0, ymax=160, ls='--', color='red')

ax2 = df.instancias.plot(secondary_y=True, label='VMs (direita)')
ax2.set_ylim(bottom=0)
#ax2.set_ylim(top=3)
plt.ylabel("Número de VMs")
plt.gca().yaxis.set_major_locator(mticker.MultipleLocator(1))

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()

plt.legend(h1+h2, l1+l2, loc='upper right')

plt.show()
```
